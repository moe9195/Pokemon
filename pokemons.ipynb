```python
import requests
import pprint
from collections import Counter
import plotly.figure_factory as ff
import pandas as pd
import operator, iterkeys
```

```python
url = "https://pokeapi.co/api/v2/pokemon/"
nums = list(range(1, 808))
urls = [url+str(num) for num in nums]
pokemons = []

for url in urls:
    response = requests.get(url)
    pokemon = response.json()
    pokemons.append(pokemon)
```

```python
df = pd.DataFrame.from_dict(pokemons)
df = df[['name', 'weight', 'height', 'base_experience']]

weights = df['weight'].values
heights = df['height'].values
experience = df['base_experience'].values
```

```python
group_labels = ['weights']
hist_data = [weights]
fig = ff.create_distplot(hist_data, group_labels, bin_size=50)
fig.update_layout(title_text='Distribution of Pokemon Weights')
fig.show()heg
```

```python
group_labels = ['heights']
hist_data = [heights]
fig = ff.create_distplot(hist_data, group_labels, bin_size=1)
fig.update_layout(title_text='Distribution of Pokemon Heights')
fig.show()
```

```python
group_labels = ['experience']
hist_data = [experience]
fig = ff.create_distplot(hist_data, group_labels, bin_size=1)
fig.update_layout(title_text='Distribution of Pokemon Base Experience')
fig.show()
```
